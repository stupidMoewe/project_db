```python
import pandas as pd
import json
```

```python
def json_cleaning(x):
    return x.replace("\"", "").replace("'", "\"").replace("True", "true").replace('False', '"false"').replace("u\"", "\"")
```

```python
business = pd.read_csv('yelp_academic_dataset_business.csv')
business.head()
```

```python
cat = business[business["categories"].notnull()]
categories = cat.categories.apply(lambda x: x.split(',')).explode().apply(lambda x: x.strip()).value_counts()
categories
```

```python
a = business.attributes.apply(lambda x: x if x != '{}' else None)
attrib_db = pd.DataFrame(a[a.notnull()])
```

```python
parsed_attributes = attrib_db.attributes.apply(json_cleaning).apply(lambda x: json.loads(x))
```

```python
# parsed_attributes = pd.DataFrame(parsed_attributes.tolist())
pd.DataFrame.from_dict(parsed_attributes.values.tolist(), orient='columns')
```

```python
tip = pd.read_csv('yelp_academic_dataset_tip.csv')
tip.head(10)
```

```python
business_id_col = pd.DataFrame.from_dict(tip.business_id.apply(json_cleaning).apply(lambda x: json.loads(x)).to_list(),orient='columns').transpose()
print("business_id done")
compliment_count_col = pd.DataFrame.from_dict(tip.compliment_count.apply(json_cleaning).apply(lambda x: json.loads(x)).to_list(),orient='columns').transpose()
print("compliment_count done")
user_id_col = pd.DataFrame.from_dict(tip.user_id.apply(json_cleaning).apply(lambda x: json.loads(x)).to_list(),orient='columns').transpose()
print("user_id done")
```

```python
def invert_guillemets(x):
    guillemet_map = {'"': "'", "'": '"'}
    return ''.join(guillemet_map.get(c, c) for c in x)

        
def spacial_json_cleaning(x):
    return pd.Series(list(x)).apply(lambda y: (y)).replace("True", "true").replace('False', '"false"').replace("u\"", "\"").replace("None", "null")
```

```python
# date_col = pd.DataFrame.from_dict(tip.date.apply(json_cleaning).apply(lambda x: json.loads(x)).to_list(), orient="columns").transpose()
# print("date done")
text_col = pd.DataFrame.from_dict(tip.text.apply(spacial_json_cleaning).apply(lambda x: json.loads(x)).to_list(),orient='columns').transpose()
print("text done")
```
